# CRISP-DM analysis
In the following paragraphs I will approach the dataset using the CRISP-DM methodology:
CRISP-DM stands for Cross-Industry Standard Process for Data Mining and is a process structured in 5 steps:
1. Business Understanding
2. Data Understanding
3. Data Preparation
4. Data Modeling
5. Results Analysis
6. Deployment (not implemented in this project)


## 1. Business Understanding

In this section, I will formulate three questions about the Stack Overflow survey dataset.

For this purpose, I will use data from the following list of years:
1. 2011

1. What languages were the most popular in each year?
2. Does the country of origin influence the preferred/mostly used language?
3. Does the amount of years in programming influence the preferred/mostly used language?
4. Does the languages currently known by a developer influence the desire to learn a specific language in the future?

## 5. Results Analysis

This section I will deepen the analysis on data results.

First question: "What languages were the most popular in each year?"

### Plotting the trends in top 10 languages throughout years
Now I will plot the trend of languages popularity throughout years.

Before plotting data, I need to make data ready to be plotted, so I will prepare a Dataframe </br>
containing only top 10 languages' by popularity, throughout reference years.

```python
s_surveys_data = {"2011": s_2011_proficiencies, "2012": s_2012_proficiencies, "2013": s_2013_proficiencies,
                  "2014": s_2014_proficiencies, "2015": s_2015_proficiencies, "2016": s_2016_proficiencies,
                  "2017": s_2017_proficiencies, "2018": s_2018_proficiencies, "2019": s_2019_proficiencies,
                  "2020": s_2020_proficiencies, "2021": s_2021_proficiencies}

languages_popularity_df = pd.DataFrame(columns=popular_languages, index=s_surveys_data.keys())
import re
for year, dataset in s_surveys_data.items():
    # retrieving data of the first 10 popular languages
    for lang in popular_languages:
        lang_re = r'\b' + re.escape(lang) + r'\Z'

        # lang_re = re.compile(r"\b" + re.escape(lang) + r"\b")
        tmp = dataset.filter(regex=lang_re)
        if len(tmp) > 0:
            languages_popularity_df.loc[year, lang] = int(tmp.values[0])

        else:
            languages_popularity_df.loc[year, lang] = 0
        del tmp
```

Now I can plot data, but before that, for readability purposes, I'll set default matplotlib size.

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
print("Before, figure default size is: ", plt.rcParams["figure.figsize"])
plt.rcParams["figure.figsize"] = (20, 10)
print("After, figure default size is: ", plt.rcParams["figure.figsize"])

```
